```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Dropout, Flatten, Dense, Activation, BatchNormalization
import tensorflow as tf
initializer = tf.keras.initializers.HeNormal()

mymodel = Sequential()

mymodel.add(Conv2D(32, (3, 3), activation='relu', input_shape=(150, 150, 3), kernel_initializer=initializer))
mymodel.add(BatchNormalization())
mymodel.add(MaxPooling2D(pool_size=(2, 2)))
mymodel.add(Dropout(0.1))

mymodel.add(Conv2D(64, (3, 3), activation='relu'))
mymodel.add(BatchNormalization())
mymodel.add(MaxPooling2D(pool_size=(2, 2)))
mymodel.add(Dropout(0.2))

mymodel.add(Conv2D(128, (3, 3), activation='relu'))
mymodel.add(BatchNormalization())
mymodel.add(MaxPooling2D(pool_size=(2, 2)))
mymodel.add(Dropout(0.3))

mymodel.add(Conv2D(256, (3, 3), activation='relu'))
mymodel.add(BatchNormalization())
mymodel.add(MaxPooling2D(pool_size=(2, 2)))
mymodel.add(Dropout(0.4))

mymodel.add(Flatten())
mymodel.add(Dense(1024, activation='relu'))
mymodel.add(BatchNormalization())
mymodel.add(Dropout(0.5))
mymodel.add(Dense(10, activation='sigmoid'))

mymodel.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

mymodel.summary()
```

# Training Inception model

```python
# Download the pre-trained weights. No top means it excludes the fully connected layer it uses for classification.
!wget --no-check-certificate \
    https://storage.googleapis.com/mledu-datasets/inception_v3_weights_tf_dim_ordering_tf_kernels_notop.h5 \
    -O /tmp/inception_v3_weights_tf_dim_ordering_tf_kernels_notop.h5
```

```python
from tensorflow.keras.applications.inception_v3 import InceptionV3
from tensorflow.keras import layers

# Set the weights file you downloaded into a variable
local_weights_file = '/tmp/inception_v3_weights_tf_dim_ordering_tf_kernels_notop.h5'

# Initialize the base model.
# Set the input shape and remove the dense layers.
pre_trained_model = InceptionV3(input_shape = (150, 150, 3),
                                include_top = False,
                                weights = None)

# Load the pre-trained weights you downloaded.
pre_trained_model.load_weights(local_weights_file)

# Freeze the weights of the layers.
for layer in pre_trained_model.layers:
  layer.trainable = False
```

```python
pre_trained_model.summary()
```

```python
# Choose `mixed_7` as the last layer of your base model
last_layer = pre_trained_model.get_layer('mixed7')
print('last layer output shape: ', last_layer.output_shape)
last_output = last_layer.output
```

```python
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import Model

# Flatten the output layer to 1 dimension
x = layers.Flatten()(last_output)
# Add a fully connected layer with 1,024 hidden units and ReLU activation
x = layers.Dense(1024, activation='relu')(x)
# Add a dropout rate of 0.2
x = layers.Dropout(0.2)(x)

x = layers.Dense(1024, activation='relu')(x)
# Add a dropout rate of 0.2
x = layers.Dropout(0.2)(x)
# Add a final sigmoid layer for classification
x = layers.Dense  (10, activation='sigmoid')(x)

# Append the dense network to the base model
model = Model(pre_trained_model.input, x)

# Print the model summary. See your dense network connected at the end.
model.summary()
```

```python
from tensorflow.keras.optimizers import Adam

model.compile(optimizer=Adam(learning_rate=0.0001), loss="binary_crossentropy", metrics=["accuracy"])
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
from keras.callbacks import EarlyStopping, ReduceLROnPlateau
earlystop = EarlyStopping(monitor='val_loss',
                          patience=10,
                          restore_best_weights=True)
learning_rate_reduction = ReduceLROnPlateau(monitor='val_loss',
                                            patience=2,
                                            verbose=1,
                                            factor=0.3,
                                            min_lr=0.00001)
callbacks = [earlystop, learning_rate_reduction]

STEP_SIZE_TRAIN = min(train_data_gen.n // train_data_gen.batch_size, 120)
STEP_SIZE_VALID = min(validation_data_gen.n // validation_data_gen.batch_size, 30)
FAST_RUN = False
```

```python
epochs=3 if FAST_RUN else 50
history=model.fit_generator(generator=train_data_gen,
                    steps_per_epoch=STEP_SIZE_TRAIN,
                    validation_data=validation_data_gen,
                    validation_steps=STEP_SIZE_VALID,
                    epochs=epochs,
                    callbacks=callbacks)
```

```python
# Plot the chart for accuracy and loss on both training and validation
import matplotlib.pyplot as plt

acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(len(acc))

plt.plot(epochs, acc, 'r', label='Training accuracy')
plt.plot(epochs, val_acc, 'b', label='Validation accuracy')
plt.title('Training and validation accuracy Inception-Net')
plt.legend()
plt.figure()

plt.plot(epochs, loss, 'r', label='Training Loss')
plt.plot(epochs, val_loss, 'b', label='Validation Loss')
plt.title('Training and validation loss')
plt.legend()

plt.show()
```

** Cũng khá ổn, không quá tệ. Accuracy giảm có thể là do dữ liệu lơn hơn và khác biệt nhiều hơn. Model mới có thể tốt hơn model với data cũ do có tính khái quát hoá cao hơn mặc dù accuracy thấp hơn

```python
def predict_test(model,image_folder,output_folder,name_model):
# Đường dẫn tới thư mục chứa ảnh cần dự đoán
  output_csv = f'{output_folder}/{name_model}.csv'
  os.makedirs(output_folder, exist_ok=True)
  # Danh sách tên các cột
  label_names = ['HG', 'HT', 'TR', 'CTH', 'BD', 'VH', 'CTQ', 'DQT', 'KS', 'CVN']

  # Mở file CSV để ghi kết quả dự đoán
  data_rows = []

  # Lặp qua tất cả các file ảnh trong thư mục
  for filename in os.listdir(image_folder):
      if filename.lower().endswith('.jpg') or filename.lower().endswith('.jpeg'):  # Đảm bảo chỉ xử lý các file ảnh
          image_path = os.path.join(image_folder, filename)

          # Đọc và xử lý ảnh
          image = Image.open(image_path)
          # Thực hiện các bước tiền xử lý ảnh (nếu cần)
          image = image.resize((150, 150))  # Đảm bảo ảnh có kích thước phù hợp với mô hình
          image = image.convert('RGB')  # Đảm bảo ảnh có 3 kênh màu (RGB)
          image_array = tf.keras.preprocessing.image.img_to_array(image)
          image_array = tf.expand_dims(image_array, 0)

          # Dự đoán nhãn ảnh
          predictions = model.predict(image_array)
          binary_predictions = np.where(predictions >= 0.5, 1, 0)

          # Ghi kết quả dự đoán vào danh sách các hàng dữ liệu
          row = [filename] + list(binary_predictions[0])
          data_rows.append(row)

  # Sắp xếp danh sách các hàng dữ liệu theo cột 'A' (tên ảnh)
  sorted_data_rows = sorted(data_rows, key=lambda x: x[0])

  # Ghi kết quả dự đoán vào file CSV
  with open(output_csv, 'w', newline='') as csvfile:
      writer = csv.writer(csvfile)
      column_names = ['A'] + label_names
      writer.writerow(column_names)
      writer.writerows(sorted_data_rows)

  print(f"{name_model} dự đoán hoàn thành và kết quả đã được lưu vào file CSV.")
```

```python
image_folder = '/content/drive/MyDrive/DeepLearning/Test'
predict_test(model_vgg_16,image_folder,image_folder,'model_vgg_16')
predict_test(model_vgg_19,image_folder,image_folder,'model_vgg_19')
predict_test(model_Inception,image_folder,image_folder,'model_Inception')
predict_test(model_rn_50,image_folder,image_folder,'model_rn_50')
predict_test(model_rn_101,image_folder,image_folder,'model_rn_101')
predict_test(model_rn_152,image_folder,image_folder,'model_rn_152')
```

```python
import pandas as pd
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python

df = pd.read_csv('/content/drive/MyDrive/Drive/DeepLearning/Anh_Va_Nhan.txt', sep = ',')
header=['A', 'HG', 'HT', 'TR', 'CTH', 'BD', 'VH', 'CTQ', 'DQT', 'KS', 'CVN']
df.columns=header
```

```python
data_gen = ImageDataGenerator(
    rotation_range=15,
    rescale=1./255, #scale giá trị của pixels từ (0, 255) về (0, 1)
    shear_range=0.1,
    zoom_range=0.2,
    horizontal_flip=True,
    width_shift_range=0.1,
    height_shift_range=0.1,
    validation_split=0.2
)

# train_data_gen = data_gen.flow_from_dataframe(directory='/content/Data',
#     target_size=(150, 150),
#     batch_size=32,
#     class_mode='raw',
#     subset='training',
#     dataframe=df[:400],
#     x_col="A",
#     y_col=header,
#     seed=42,
#     shuffle=True
# )

train_data_gen=data_gen.flow_from_dataframe(
    dataframe=df[:400],
    directory="/content/model",
    x_col='A',
    y_col=columns,
    batch_size=32,
    seed=42,
    shuffle=True,
    class_mode="raw",
    target_size=(150,150))

validation_data_gen=data_gen.flow_from_dataframe(
    dataframe=df[400:],
    directory="/content/model",
    x_col='A',
    y_col=columns,
    batch_size=32,
    seed=42,
    shuffle=True,
    class_mode="raw",
    target_size=(150,150))
# validation_data_gen = data_gen.flow_from_dataframe(directory='/content/Data',
#     target_size=(150, 150),
#     batch_size=32,
#     class_mode='raw',
#     subset='validation',
#     dataframe=df[401:],
#     x_col="A",
#     y_col=header,
#     seed=42,
#     shuffle=True)
```

```python
# Download the pre-trained weights. No top means it excludes the fully connected layer it uses for classification.
!wget --no-check-certificate \
    https://storage.googleapis.com/mledu-datasets/inception_v3_weights_tf_dim_ordering_tf_kernels_notop.h5 \
    -O /tmp/inception_v3_weights_tf_dim_ordering_tf_kernels_notop.h5
```

```python
from tensorflow.keras.applications.inception_v3 import InceptionV3
from tensorflow.keras import layers

# Set the weights file you downloaded into a variable
local_weights_file = '/tmp/inception_v3_weights_tf_dim_ordering_tf_kernels_notop.h5'

# Initialize the base model.
# Set the input shape and remove the dense layers.
pre_trained_model = InceptionV3(input_shape = (150, 150, 3),
                                include_top = False,
                                weights = None)

# Load the pre-trained weights you downloaded.
pre_trained_model.load_weights(local_weights_file)

# Freeze the weights of the layers.
for layer in pre_trained_model.layers:
  layer.trainable = False
```

```python
# Choose `mixed_7` as the last layer of your base model
last_layer = pre_trained_model.get_layer('mixed7')
print('last layer output shape: ', last_layer.output_shape)
last_output = last_layer.output
```

```python
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import Model

# Flatten the output layer to 1 dimension
x = layers.Flatten()(last_output)
# Add a fully connected layer with 1,024 hidden units and ReLU activation
x = layers.Dense(1024, activation='relu')(x)
# Add a dropout rate of 0.2
x = layers.Dropout(0.2)(x)
# Add a final sigmoid layer for classification
x = layers.Dense  (10, activation='sigmoid')(x)

# Append the dense network to the base model
model = Model(pre_trained_model.input, x)

# Print the model summary. See your dense network connected at the end.
model.summary()
```

```python
from tensorflow.keras.optimizers import Adam

model.compile(optimizer=Adam(learning_rate=0.0001), loss="binary_crossentropy", metrics=["accuracy"])
```

```python
from keras.callbacks import EarlyStopping, ReduceLROnPlateau
earlystop = EarlyStopping(monitor='val_loss',
                          patience=10,
                          restore_best_weights=True)
learning_rate_reduction = ReduceLROnPlateau(monitor='val_loss',
                                            patience=2,
                                            verbose=1,
                                            factor=0.3,
                                            min_lr=0.00001)
callbacks = [earlystop, learning_rate_reduction]

STEP_SIZE_TRAIN = min(train_data_gen.n // train_data_gen.batch_size, 120)
STEP_SIZE_VALID = min(validation_data_gen.n // validation_data_gen.batch_size, 30)
FAST_RUN = False
```

```python
epochs=3 if FAST_RUN else 50
history=model.fit_generator(generator=train_data_gen,
                    steps_per_epoch=STEP_SIZE_TRAIN,
                    validation_data=validation_data_gen,
                    validation_steps=STEP_SIZE_VALID,
                    epochs=epochs,
                    callbacks=callbacks)
```

```python
# Plot the chart for accuracy and loss on both training and validation
import matplotlib.pyplot as plt

acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(len(acc))

plt.plot(epochs, acc, 'r', label='Training accuracy')
plt.plot(epochs, val_acc, 'b', label='Validation accuracy')
plt.title('Training and validation accuracy Inception-Net in private Dataset')
plt.legend()
plt.figure()

plt.plot(epochs, loss, 'r', label='Training Loss')
plt.plot(epochs, val_loss, 'b', label='Validation Loss')
plt.title('Training and validation loss')
plt.legend()

plt.show()
```

```python
import pandas as pd

def calculate_accuracy(predicted_labels_df, actual_labels_df):
    # Lấy danh sách tên ảnh từ hai DataFrame
    predicted_labels = predicted_labels_df.iloc[:, 0].str.extract(r'(\d+)').astype(int)
    actual_labels = actual_labels_df.iloc[:, 0].str.extract(r'(\d+)').astype(int)

    # Sắp xếp theo tên ảnh
    predicted_labels_df['Label'] = predicted_labels
    predicted_labels_df = predicted_labels_df.sort_values(by='Label')
    predicted_labels_df.reset_index(drop=True, inplace=True)

    actual_labels_df['Label'] = actual_labels
    actual_labels_df = actual_labels_df.sort_values(by='Label')
    actual_labels_df.reset_index(drop=True, inplace=True)

    # Đếm số lượng nhãn dự đoán khớp với nhãn thực tế
    correct_predictions = (predicted_labels_df.iloc[:, 1:] == actual_labels_df.iloc[:, 1:]).all(axis=1).sum()

    # Tính tổng số dòng trong DataFrame
    total_samples = predicted_labels_df.shape[0]

    # Tính độ chính xác
    accuracy = correct_predictions / total_samples

    return accuracy
```
